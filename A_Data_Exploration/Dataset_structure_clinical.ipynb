### So you get an idea on how to merge the two datasets from TCGA microbial and clinical, the following steps perform merging of the data trough slicing only for the first 14 characters

```python
micro = pd.read_excel("/Users/laurinseelig/Desktop/ZHAW Semester 5/PA2/Data Set/Data/TCGA_Microbial_Content/data_file_s12.xlsx")
```

```python
micro
```

Tumor_Sample_Barcode and alquot_submitter_id are beeing used to merge the data due to being the same overlapping instances

```python
# Extract only the first 15 characters (including dashes)
clin["merge_id"] = clin["Tumor_Sample_Barcode"].astype(str).str.slice(0, 14)
micro["merge_id"] = micro["aliquot_submitter_id"].astype(str).str.slice(0, 14)

# Optional: check a few examples
print("Clin example IDs:", clin["merge_id"].head(3).tolist())
print("Micro example IDs:", micro["merge_id"].head(3).tolist())

# Merge on the new standardized key
merged = pd.merge(micro, clin, on="merge_id", how="inner")

print(f"Merged rows: {merged.shape[0]}")
print(f"Columns in merged: {merged.shape[1]}")
merged
```
